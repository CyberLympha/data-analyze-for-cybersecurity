# Прогнозирование временных рядов с помощью моделей LSTM

Датасет "data_example.csv" (Практика 2.1)

**Задача**

Для холодильной горки "18СТ" (или любой другой) реализовать следующие модели:

0. Для одномерного и многомерного прогнозирования напишите функции для скользящего окна.
1. Одномерное одношаговое прогнозирование температуры воздуха на входе в испаритель и внутри установки.
2. Многомерное одношаговое прогнозирование температуры воздуха на входе в испаритель и внутри установки (с одним выходом и с двумя выходами).
3. Одномерное многошаговое (несколько минут и часов) прогнозирование температуры воздуха на входе в испаритель и внутри установки.
4. Многомерное многошаговое (несколько минут и часов) прогнозирование температуры воздуха на входе в испаритель и внутри установки (с одним выходом и с двумя выходами).
5. Для каждой модели вычислите метрики качества.
6. Ипользуя метрики качества, для каждой модели LSTM сделайте сравнение с предсказаниями стат.модельи ARIMA.

Для пунктов 1-4 нужно определить гиперпараметры, подобрать оптимальное количество слоев НС.
Определить количество эпох обучения (из графика кривой обучения).

**Важно!** Нужно разделить датасет на обучающую и тестовую выборку.
